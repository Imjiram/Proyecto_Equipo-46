## Definiciones

- **Imágenes de Calcio:** Técnicas de imagen que utilizan indicadores sensibles al calcio para visualizar y medir la actividad neuronal a través de cambios en los niveles de calcio intracelular.
- **Neurona:** Célula del sistema nervioso cuya función principal es recibir, procesar y transmitir información a través de señales eléctricas y químicas.
- **Potencial de Acción:** Breve inversión del potencial eléctrico de la membrana de una neurona, que permite la transmisión de la señal a lo largo del axón.
- **Sinapsis:** Estructura que permite a una neurona transmitir una señal eléctrica o química a otra célula.
- **Microscopía de Dos Fotones:** Técnica de microscopía que permite la visualización de tejidos vivos hasta una profundidad mayor que la microscopía de fluorescencia convencional, utilizando dos fotones de baja energía.
- **Fluorescencia:** Emisión de luz por una sustancia que ha absorbido luz u otra radiación electromagnética. Es una forma común de visualizar procesos biológicos.
- **Indicador de Calcio:** Molécula que se une al calcio y cambia sus propiedades ópticas, permitiendo la detección de cambios en la concentración de calcio dentro de las células.
- **Actividad Neuronal:** Conjunto de procesos electroquímicos que se realizan en las neuronas, incluyendo la transmisión de impulsos nerviosos.
Conectividad Neuronal: Refiere a cómo las neuronas están enlazadas entre sí y cómo estas conexiones contribuyen al procesamiento de información.
- **Evento Neuronal:** Actividad registrada de una neurona que generalmente incluye disparos o potenciales de acción, indicativa de procesamiento de información.

## Raster Plots


**Definición:** Un raster plot es un tipo de gráfico utilizado en neurociencia para representar eventos neuronales a lo largo del tiempo. Cada disparo de una neurona se muestra como un punto o una marca en una línea que representa el tiempo, con cada línea horizontal del gráfico correspondiendo a una neurona diferente o al mismo neurón en ensayos repetidos.

**Propósito:** Los raster plots son utilizados para visualizar y analizar la actividad temporal de las neuronas durante diferentes tipos de estímulos o comportamientos. Permiten a los investigadores observar patrones de disparo, como la sincronización y la frecuencia de los potenciales de acción.

**Análisis de Patrones:** Al examinar los raster plots, los científicos pueden identificar correlaciones entre la actividad neuronal y estímulos externos, así como coordinaciones entre diferentes neuronas, lo que es crucial para entender procesos como la percepción sensorial y la toma de decisiones.
Aplicaciones: Son especialmente útiles en experimentos donde la respuesta temporal precisa de las neuronas es crítica, como en estudios de decisiones motoras o procesamiento sensorial.

## Ensamblajes Neuronales

**Definición:** Un ensamble neuronal se refiere a un grupo de neuronas que se activan juntas en un patrón repetitivo y coherente. Estos ensambles suelen interpretarse como las unidades funcionales básicas del procesamiento de información en el cerebro.

**Formación de Ensamblajes:** Los ensamblajes neuronales pueden formarse en respuesta a estímulos específicos o durante ciertas fases del comportamiento o el aprendizaje. La formación de estos ensambles está influenciada por la plasticidad sináptica, que es la capacidad de las conexiones entre neuronas para cambiar en fuerza y eficacia.
Rol en el Procesamiento Cerebral: Se cree que los ensamblajes neuronales juegan un papel crucial en funciones cognitivas como la memoria, el aprendizaje y la percepción. La actividad coordinada de un ensamble puede representar un componente específico de una experiencia sensorial o un recuerdo.

**Detección y Análisis:** Identificar ensamblajes neuronales a menudo requiere métodos estadísticos avanzados que pueden detectar patrones de actividad que se repiten en el tiempo y entre diferentes células. Estos métodos incluyen análisis de correlación, técnicas de clustering, y algoritmos de aprendizaje de máquina.
Implicaciones Clínicas: El estudio de ensamblajes neuronales puede tener implicaciones en el entendimiento de desórdenes neurológicos, donde la actividad neuronal normal puede estar alterada. Por ejemplo, en enfermedades como la epilepsia, los patrones de ensamble pueden ser anormalmente sincronizados y generalizados.

Este primer avance consiste en realizar un análisis exploratorio de datos (EDA - Exploratory Data Analysis), es decir, describir los datos utilizando técnicas estadísticas y de visualización (análisis univariante y bi/multivariante) para hacer enfoque en sus aspectos más relevantes, así como aplicar y justificar operaciones de preprocesamiento, relacionadas con el manejo de valores faltantes, atípicos y alta cardinalidad.

```python
import cv2
import numpy as np
from google.colab import files
from matplotlib import pyplot as plt
```

```python
uploaded = files.upload()

nombre_archivo = list(uploaded.keys())[0]

# Leer la imagen usando OpenCV
imagen = cv2.imread(nombre_archivo)

# Convertir la imagen a escala de grises
imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

# Normalizar la imagen
imagen_normalizada = cv2.normalize(imagen_gris, None, 0, 255, cv2.NORM_MINMAX)

# Aplicar detección de bordes con el algoritmo Canny
bordes = cv2.Canny(imagen_normalizada, 50, 150)

# Aplicar la transformada de Hough para detectar líneas
lineas = cv2.HoughLinesP(bordes, 1, np.pi/180, threshold=50, minLineLength=100, maxLineGap=5)

# Dibujar las líneas detectadas sobre los bordes
imagen_con_lineas = cv2.cvtColor(bordes, cv2.COLOR_GRAY2BGR)
if lineas is not None:
    for linea in lineas:
        x1, y1, x2, y2 = linea[0]
        cv2.line(imagen_con_lineas, (x1, y1), (x2, y2), (0, 255, 0), 2)

# Mostrar la imagen original, los bordes y las líneas detectadas
plt.figure(figsize=(15,5))

plt.subplot(1, 3, 1)
plt.title('Imagen Original')
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Bordes Detectados (Canny)')
plt.imshow(bordes, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Líneas Detectadas (Hough)')
plt.imshow(cv2.cvtColor(imagen_con_lineas, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```
